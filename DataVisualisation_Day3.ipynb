```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('fmri')
data
```

```python
sns.relplot(x='timepoint', y = 'signal',data = data,hue = 'event')
```

```python
sns.lineplot(x='signal',y = 'region',data = data,size = 'subject')
```

```python
sns.boxplot(x='subject',y ='timepoint',hue ='region',data = data )
```

```python
sns.catplot(x= 'signal',y = 'subject',kind = 'violin',data = data)
```
